# Are there any noticeable trends in February temperatures over time for the various locations?

# A trend I see in February temperatures over time for the various locations is that the average temperature in the Midwest is higher than the northeast or Pennsylvania

```python
plt.scatter(data['Year'], data['February Average Temperature'], label = 'February')
plt.scatter(data['Year'], data['March Average Temperature'], label = 'March')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.legend()
```
